Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
import pandas as pd

def check_region(keyword):
    geo_data = { 'Центр': ['москва', 'тула', 'ярославль'],
                 'Северо-Запад': ['петербург', 'псков', 'мурманск'],
                 'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']  }
    for key, values in geo_data.items():
        for value in values:
            if  value in keyword:
                return key
    return 'undefined'


keywords = pd.read_csv('keywords.csv')
keywords['region'] = keywords['keyword'].apply(check_region)
keywords.head()
```


**Задание 2**

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
    - оценка 2 и меньше - низкий рейтинг
    - оценка 4 и меньше - средний рейтинг
    - оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

Так как в задании не учтены оценки (4, 4.5), то были приняты следующие правила:

x ∈ [0, 2] - низкий рейтинг

x ∈ (2, 4] - средний рейтинг

x ∈ (4, 5] - высокий рейтинг


```python
import pandas as pd

def class_movie(x):
    if x > 4:
        return 'высокий рейтинг'
    elif x > 2:
        return 'средний рейтинг'
    else:
        return 'низкий рейтинг'

movies = pd.read_csv('ml-latest-small/movies.csv') 
ratings = pd.read_csv('ml-latest-small/ratings.csv')

ratings = ratings[['movieId', 'rating']].groupby('movieId').mean()
ratings['class'] = ratings['rating'].apply(class_movie)
ratings = ratings.merge(movies, on='movieId', how='left')   
ratings[['movieId', 'title', 'genres', 'rating', 'class']]
```

**Задание 3**

Посчитайте среднее значение Lifetime киноманов (пользователи, которые поставили 100 и более рейтингов). Под Lifetime понимается разница между максимальным и минимальным значением timestamp для каждого пользователя. Ответ дайте в днях.


```python
# создание датафрейма с данными по всем записям о выставлении пользователями рейтингов
import pandas as pd
from datetime import datetime
ratings = pd.read_csv('ml-latest-small/ratings.csv')                                    
```

```python
# группировка по пользователю и выявление только тех, число оценок (строк) для которых >= 100 в датафрейм ratings_filtered
ratings_filtered = ratings.groupby('userId').count()             
ratings_filtered = ratings_filtered[ratings_filtered['timestamp'] >= 100].reset_index()
ratings_filtered.head(5)
```

```python
#в датафрейме ratings оставляются только оценки выбранных пользователей
#данные группируются по userid с применением аггрегирующих ф-й min и max к timestamp, 
#после чего вычисляется разность этих чисел в днях и записывается в поле lifetime

ratings = ratings[ratings['userId'].isin(ratings_filtered['userId'].tolist())]
ratings = ratings.groupby('userId').agg(['min', 'max'])['timestamp']
ratings['lifetime'] = ratings\
            .apply(lambda x: abs(datetime.fromtimestamp(x['max']) - datetime.fromtimestamp(x['min'])).days, axis=1)
ratings.head(20)
```

```python
#вычисляется среднее кол-во дней по таблице

mean_lifetime = ratings.agg(['mean'])['lifetime']
print(mean_lifetime.values[0])
```

**Задание 4**

Есть мнение, что "раньше снимали настоящее кино, не то что сейчас". Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из материалов занятия. Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
    - для каждой строки пройдите по всем годам списка years
    - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year'

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца 'year' и отсортируйте результат по убыванию рейтинга

```python
import pandas as pd

def production_year(raw):                                       # функция возвращает число (год) от 1950 до 2010,
    for x in range(1950,2011):                                  # если он встерчается в названии фильма, иначе - 1900
        if str(x) in raw['title']:                       
            return x
    return 1900

movies = pd.read_csv('ml-latest-small/movies.csv')              # создается датафрейм movies
movies['year'] = movies.apply(production_year, axis = 1)        # добавляется новое поле year при помощи функции production_year
ratings = pd.read_csv('ml-latest-small/ratings.csv')            # создается датафрейм ratings
joined = ratings.merge(movies, on='movieId', how='left')        # два датафрейма объединяются по movieId
joined = joined.groupby('year').agg('mean')['rating']           # вычисляется средние значения rating при группировке по year
print(joined.sort_values(ascending=False))                      # сортировка и вывод
```
