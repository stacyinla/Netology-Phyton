geo_logs = [

ids = {'user1': [213, 213, 213, 15, 213], 'user2': [54, 54, 119, 119, 119], 'user3': [213, 98, 98, 35]}

**Задание 3**

Список поисковых запросов. Получить распределение количества слов в них. Т. е. поисковых запросов из одного слова 5%, из двух - 7%, из трех - 3% итд.

queries = [
'смотреть сериалы онлайн',
'новости спорта',
'афиша кино',
'курс доллара',
'сериалы этим летом',
'курс по питону',
'сериалы про спорт',
]

```python
#решение состоит в переборе элементов списка и формировании словаря, где ключ - кол-во слов в элементе списка, а в значении
#суммируется кол-во элементов с такой длиной

queries = ['смотреть сериалы онлайн', 'новости спорта', 'афиша кино', 'курс доллара', 'сериалы этим летом', 
           'курс по питону', 'сериалы про спорт',]
words_count = dict()

procent = 1 / len(queries)
for x in queries:
    if len(x.split()) in words_count:
        words_count[len(x.split())] += procent
    else:
        words_count[len(x.split())] = procent
        
for key in words_count:
    print("строк из {:d} слов - {:.4%}".format(key,words_count[key]))
    
    
```

**Задание 4**

Дана статистика рекламных каналов по объемам продаж. Напишите скрипт, который возвращает название канала с максимальным объемом.

Т. е. в данном примере скрипт должен возвращать 'yandex'.

stats = {'facebook': 55, 'yandex': 120, 'vk': 115, 'google': 99, 'email': 42, 'ok': 98} 

```python
#решение состоит в переборе элементов словаря и фиксации значения, большего, чем было зафиксировано ранее
#изначально показатель сравнивается с counter == 0, т.к. по заданию значение неотрицательно (объем не мб отрицательным)

stats = {'facebook': 55, 'yandex': 120, 'vk': 115, 'google': 99, 'email': 42, 'ok': 98}
counter = 0
max_channel = ''

for key in stats:
    if stats[key] > counter:
        counter = stats[key]
        max_channel = key
        
print(max_channel)
```

**Задание 5**

Дан поток логов по количеству просмотренных страниц для каждого пользователя. Список отсортирован по ID пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей. Учтите, что весь список stream не помещается в оперативную память, т. е. его нужно обрабатывать поэлементно в цикле.

stream = ['2018-01-01,user1,3', '2018-01-07,user1,4', '2018-03-29,user1,1', '2018-04-04,user1,13', '2018-01-05,user2,7', '2018-06-14,user3,4', '2018-07-02,user3,10', '2018-03-21,user4,19', '2018-03-22,user4,4', '2018-04-22,user4,8', '2018-05-03,user4,9', '2018-05-11,user4,11',]

```python
#решение состоит в переборе элементов списка (строк), разделении их по разделителю (,) и формировании словаря, где ключом 
#является наименование пользователя, а в значении суммируется число его просмотров. 
#Уникальное число пользователей определяется как число записей в словаре.
#Обращение к списку - по номеру элемента, который подается с помощью range, то есть по одному элементу в памяти.

user_review = dict()
stream = [
'2018-01-01,user1,3',
'2018-01-07,user1,4',
'2018-03-29,user1,1',
'2018-04-04,user1,13',
'2018-01-05,user2,7',
'2018-06-14,user3,4',
'2018-07-02,user3,10',
'2018-03-21,user4,19',
'2018-03-22,user4,4',
'2018-04-22,user4,8',
'2018-05-03,user4,9',
'2018-05-11,user4,11',
]
for x in range(len(stream)): 
    if stream[x].split(',')[1] in user_review:                                  #если id юзера уже в словаре
        user_review[stream[x].split(',')[1]] += int(stream[x].split(',')[2])    #плюсуем ему просмотры
    else:                                                                       #иначе -
        user_review[stream[x].split(',')[1]] = int(stream[x].split(',')[2])     #добавляем ключ и соотв.знач. просмотров
        
print([[key, user_review[key]/len(user_review)] for key in user_review])        #вывод с помощью list comprehension
    
```
