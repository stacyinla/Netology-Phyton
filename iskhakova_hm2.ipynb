**Задание 3**

Список поисковых запросов. Получить распределение количества слов в них. Т. е. поисковых запросов из одного слова 5%, из двух - 7%, из трех - 3% итд.

queries = [
'смотреть сериалы онлайн',
'новости спорта',
'афиша кино',
'курс доллара',
'сериалы этим летом',
'курс по питону',
'сериалы про спорт',
]

```python
queries = ['смотреть сериалы онлайн', 'новости спорта', 'афиша кино', 'курс доллара', 'сериалы этим летом', 
           'курс по питону', 'сериалы про спорт',]
words_count = dict()

procent = 1 / len(queries)
for x in queries:
    if len(x.split()) in words_count:
        words_count[len(x.split())] += procent
    else:
        words_count[len(x.split())] = procent
        
for key in words_count:
    print("строк из {:d} слов - {:.4%}".format(key,words_count[key]))
    
    
```
