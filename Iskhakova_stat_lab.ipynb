```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```


***1. Загрузить файл auto.csv***


```python
auto = pd.read_csv('auto.csv')
auto.head()
```


***2. Рассчитать основные статистики для переменных(среднее, медиана, мода, мин/макс, сред. отклонение)***


```python
auto.describe()
```

```python
engine_size = auto['engine-size']
engine_size.mean()                            #среднее значение по показателю engine-size
```

```python
engine_size.median()                         # медиана значений по показателю engine-size
```

```python
engine_size.mode().tolist()                  # мода значений по показателю engine-size
```

Два значения, так как и 90 и 122 встречается равное кол-во раз (15)

```python
engine_size.min()                            # минимальное значение в столбце engine-size
```

```python
engine_size.max()                            # максимальное значение в столбце engine-size
```

```python
np.var(engine_size)                          # дисперсия
```

```python
np.std(engine_size)                          # среднеквадратичное отклонение
```

```python
auto['engine-size'].describe()              # проверю себя с помощью describe()
```


***3. Рассчитать и визуализировать корреляционную матрицу для переменных***


***4. Визуализировать распределение переменной ‘Price’, рассчитать показатели skewness и kurtosis***


```python
plt.hist(auto['price'], 60)
```

```python
sts.kurtosis(auto['price'])
```

```python
sts.skew(auto['price'])
```


***5. Визуализировать несколько взаимосвязей между переменными***


Найдем и визуализируем взаимосвязь между engine-size и price.

Как видно, имеется значительная корреляция этих двух выборок.

Можно сделать вывод, что значения engine-size  влияют на цену автомобиля.

```python
sts.pearsonr(auto['engine-size'], auto['price'])[0]
```

```python
plt.plot(auto['engine-size'], auto['price'], 'r.')
```

Найдем и визуализируем корреляцию между normalized-losses и price.

Как видно, корреляция этих двух выборок очень низкая. 

Можно сделать вывод, что значения normalized-losses не влияют на цену.

```python
sts.pearsonr(auto['normalized-losses'], auto['price'])[0]
```

```python
plt.plot(auto['normalized-losses'], auto['price'], 'r.')
```
