```python
import pandas as pd
import matplotlib.pyplot as plt
```

*****1. Распределение количества игроков NBA по числу лет, проведенных в игре*****

Тип графика - гистограмма

```python
players = pd.read_csv("nba-players-stats/player_data.csv")
players["years"] = players["year_end"] - players["year_start"]
players.head()
```

```python
players = players[["years", "name"]].groupby("years").count().reset_index()
players
```

```python
plt.barh(players['years'], players["name"])
plt.xlabel('Quantity of players')
plt.ylabel('Years of participation')
plt.title('The duration of participation of players in basketball')
plt.grid()
```

Переменные числового типа.

Цель - увидеть, какое количество лет играют спортсмены в баскетбол.
 
Как видно по гистограмме - подавляющее большинство игроков были в игре менее 1 года. Дальнейшее убывание почти во всех случаях закономерно, кроме уровня в 9 и 10 лет.
Резкий спад (уход игроков) - почти в 2 раза - наблюдается после 13 лет в игре.

****4. Распределение возраста начала спортивной карьеры игроков NBA****

```python
import seaborn as sns
```

```python
def cut_year(date):
    date = str(date)[-4:]
    try:
        return int(date)
    except ValueError:
        return 1930



players = pd.read_csv("nba-players-stats/player_data.csv")
players["birth_year"] = players["birth_date"].apply(cut_year)

players["age_start"] = players["year_start"] - players["birth_year"]
players.head(20)
```

```python
players = players[["age_start", "name"]].groupby("age_start").count().reset_index()
players
```

```python
sns.distplot(players.age_start)
```

Тип данных - числовой.

Цель - увидеть распределение числового показателя возраста начала спортивной карьеры игроков NBA.
